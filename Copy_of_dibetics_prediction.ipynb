```python
# training and normalization of data
X = data.iloc[:,:8].values #independent variable
Y = data.iloc[:,8:].values #dependent variable


X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.20, random_state=1)

# SScaler = StandardScaler()
# X_train = SScaler.fit_transform(X_train)
# X_test = SScaler.fit_transform(X_test)
```

```python
# # Fitting Linear Regression to the dataset
# from sklearn.linear_model import LinearRegression
# lr = LinearRegression()
# lr.fit(X_train,Y_train)
```

```python
from sklearn.linear_model import LogisticRegressionCV
model = LogisticRegressionCV(cv=8, max_iter=800)
model.fit(X_train,Y_train)
prediction = model.predict(X_test)
print(prediction)
```

```python
test_data = data.iloc[600:601]
output = test_data['Outcome']
test_data.drop(['Outcome'], axis =1 , inplace = True)
print("Predicted output : ",model.predict(test_data)[0])
print("Actual output : ",output.values[0])
```

```python
print("Assuracy score = ", model.score(X_train, Y_train))
```

```python
print("Train accuracy score = ", round(model.score(X_train, Y_train)*100,2))
print("Test accuracy score = ", round(model.score(X_train, Y_train)*100,2))
```
